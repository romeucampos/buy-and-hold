```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
data3 = pd.read_csv('BTC_BRL MercadoBitcoin.csv', thousands='.', decimal=',')
data2 = pd.read_csv('BTC_BRL MercadoBitcoin.csv',thousands='K',  decimal=',')


data = pd.concat([data2['Data'], data3[['Abertura','Máxima','Mínima','Último', 'Vol.']]], axis=1) 

data['Data']=pd.to_datetime(data['Data'], format="%d.%m.%Y")
data.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume']
data['Volume'] = data['Volume'].str.replace('K', ' ').str.replace(',','.').astype(float).mul(1000)
data.set_index('Date', inplace=True)
data
```

```python
data.to_csv(f'BTC-BRL-{str(data.index[-1]).split(" ")[0]}.csv')
```
